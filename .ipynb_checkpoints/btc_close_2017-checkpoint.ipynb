````python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import pygal
````

````python
df = pd.read_json('btc_close_2017.json')
df.date = df.date.dt.date

df.head()
````

````python
line_chart = pygal.Line(x_label_rotation=20, show_minor_x_labels=False)

line_chart.title = ''
line_chart.x_labels = df.date
N = 20 # we will plot only every 5 date from date_list
line_chart.x_labels_major = list(df.date)[::N]
line_chart.add('', df.close)
````

````python
line_chart.render_to_file('btc_close_2017.svg')
````
