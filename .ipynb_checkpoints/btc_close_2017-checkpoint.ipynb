````python
import json
# 
filename = 'btc_close_2017.json'
with open(filename) as f: 
    btc_data = json.load(f) # 1
# 12
for btc_dict in btc_data: # 2
    if btc_dict['month'] == '12': # 3
        date = btc_dict['date'] # 4
        week = btc_dict['week']
        weekday = btc_dict['weekday']
        close = btc_dict['close']
        print("{} is week {}, {}, the close price is {} RMB".format(date, week, weekday, close))
````
